```python
import cv2
import glob
import os

# Membuat folder output jika belum ada
# output_folder = r'C:\Users\User\Downloads\sibi\compressed\A'
# os.makedirs(output_folder, exist_ok=True)
# os.mkdir('C:\Users\User\compressed\A')

output_folder = r'C:\Users\User\Downloads\sibi\A\compressed\A' 

img_paths = glob.glob('C:\Users\User\Downloads\sibi\A\*.jpg')

# Menghitung jumlah item dalam img_paths
print(f'Jumlah Item : {len(img_paths)}')

i = 0
for img_path in img_paths:
    img = cv2.imread(img_path)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow('Gray Images', gray_img)
    output_path = os.path.join(output_folder, f'image_A_{i}.jpg')
    cv2.imwrite(output_path, gray_img)
    i += 1
    cv2.waitKey(600)
    cv2.destroyAllWindows()

print("Proses kompres selesai!")
```
